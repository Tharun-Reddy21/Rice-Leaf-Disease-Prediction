```python
import os
image_list, label_list = [], []

test_root='C:/Users/biyya/min prjct/final/test224'
root_dir = os.listdir(test_root)
for directory in root_dir :
    # remove .DS_Store from list
    if directory == ".DS_Store" :
        root_dir.remove(directory)

for disease_folder in root_dir :
    plant_disease_image_list = os.listdir(f"{test_root}/{disease_folder}")

    for image in plant_disease_image_list:
        image_directory = f"{test_root}/{disease_folder}/{image}"
        if image_directory.endswith(".jpg") == True or image_directory.endswith(".JPG") == True:
            image_list.append(image_directory)
            label_list.append(disease_folder)
print("[INFO] Image loading completed")
```

```python
import cv2
import os
# Utility
import itertools
import random
from collections import Counter
from glob import iglob


def load_image(filename):
    img = cv2.imread(filename)
    img = cv2.resize(img, (256, 256) )
    img = img /255
    
    return img


def predict(image):
    probabilities = model.predict(np.asarray([img]))[0]
    class_idx = np.argmax(probabilities)
    return probabilities
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
# print(newpath)
actuallabs=[]
predictedlabs=[]
newpath='C:/Users/biyya/min prjct/final/test'
for i in range(len(image_list)):
    filename=image_list[i]
    filename=filename.replace('\\','/')
    
# diseases=['Bacterialblight' ,'Blast' ,'Brownspot' ,'Healthy']    
    if 'Bacterialblight' in filename:
        actuallabs.append(0)
    elif 'Blast' in filename:
        actuallabs.append(1)
    elif 'Brownspot' in filename:
        actuallabs.append(2)
    elif 'Healthy' in filename:
        actuallabs.append(3)
    img = load_image(filename)
    # print(img.shape())
    prediction =  np.argmax(predict(img))
    predictedlabs.append(prediction)
    # print(diseases[prediction])
    # # print("PREDICTED: class: " , (prediction))
    # plt.imshow(img)
    # # plt.figure(idx)    
    # plt.show()
ac=np.array(actuallabs)
pr=np.array(predictedlabs)
```
